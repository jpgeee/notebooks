## 4. Branchements conditionnels

Un branchement conditionnel est une instruction qui teste une affirmation, et suivant si elle est juste ou fausse, exécute différentes instructions regroupées en un *bloc*. 
En pseudo-code, on la notera ainsi:
<pre>
<b>si</b> condition <b>alors</b>
    instruction 1
    instruction 2
    ...
<b>sinon</b>
    ...
<b>fin</b> <b>si</b>
</pre>


En Python, un bloc se manifeste par une *indentation*, c'est-à-dire un certains nombre d'espaces en début de ligne. Quand ces espaces disparaissent, le bloc est terminé. En Python, un bloc est toujours annoncé par un `:` en fin de ligne précédente.
La syntaxe d'un test conditionnel est la suivante:
```python
if test:
    instructions du bloc 1
else:
    instructions du bloc 2
```

qui exécute le bloc 1 si le test est vrai, et le bloc 2 sinon. La partie avec le `else` est facultative. 


Le tableau suivant rappelle les principaux tests en Python:


--------------------
--------------------
   `>`,`>=`        supérieur strict, supérieur ou égal     
   `<`,`<=`        inférieur strict, inférieur ou égal     
    `==`         égal     
    `!=`         différent     

--------------------
--------------------



Plusieurs conditions peuvent être combinées avec `and` ou `or`.


Par exemple, voici un programme qui teste si un nombre `x` est positif:
```python
if x>0:
    print("x est strictement positif")
else:
    print("x est négatif ou nul")
```
*Remarque*: le résultat d'un test est un *booléen*, qui peut prendre 2 valeurs: `True` ou `False`.

## 5. Boucles

Les *boucles* permettent de répéter un bloc plusieurs fois. Il existe deux grands types de boucles:
* les boucles **pour** permettent de répéter un bloc d'instructions un nombre de fois connu à l'avance:
<pre>
<b>pour</b> variable <b>allant de</b> ... <b>à</b> ... <b>faire</b>
    instruction 1
    ...
<b>fin</b> <b>pour</b>
</pre>

En Python, ce type de boucle se réalise avec le mot-clé `for`: un *itérateur* parcourt un certain nombre de valeurs choisies à l'avance; le plus souvent, ces valeurs sont générées avec la fonction `range(debut,fin,pas)` qui génère la liste des valeurs entre `debut` (inclus) et `fin` (**exclu**) par le pas donné (le pas est facultatif, par défaut il vaut 1). Exemple qui affiche les entiers entre 1 et 9:

```python
for i in range(1,10): 
    print(i)
```

* les boucles **tant que** permettent de répéter un bloc d'instructions en fonction d'une certaine condition; attention, si elle est mal programmée, une telle boucle peut tourner à l'infini:
<pre>
<b>tant que</b> condition <b>faire</b>
    instruction 1
    ...
<b>fin</b> <b>tant que</b>
</pre>

En Python, le mot-clé est `while`. Exemple qui affiche toutes les puissances de 2 inférieures à 1000:
```python
valeur=1
while valeur<1000:
    print(valeur)
    valeur=valeur*2
```



Les instructions de début de boucle sont aussi terminées par un `:` et suivies d'un bloc indenté.

## 7. Les fonctions

Une fonction est un outil pour ne pas reécrire plusieurs fois les mêmes instructions: on les place dans un bloc auquel on donne un nom, et on n'a plus qu'à appeler ce nom pour que les instructions s'exécutent. Une fonction est caractérisée par trois choses:
* ses *arguments*, c'est-à-dire les valeurs qu'il faut lui fournir 
* son *corps*, qui contient ses instructions
* sa ou ses *valeurs(s) retournée(s)*, autrement dit sa réponse.

Voici un exemple de fonction:
```python
def maximum(a,b):
    """ Cette fonction calcule le maximum entre deux réels
    arguments: a: un réel
               b: un réel
    renvoie: le plus grand des deux réels """
    if a>b:
        return a
    else:
        return b
```

Analysons cette fonction ligne par ligne:
* `def maximum(a,b):` définit la fonction par le mot-clé `def` suivi du nom de la fonction; entre parenthèses, on a une liste des arguments de la fonction: cette fonction nécessite qu'on lui communique 2 variables.
* On a ensuite des lignes de commentaires (soit avec des `#`, soit avec la syntaxe `"""` spéciale pour les fonctions) qui expliquent ce qu'elle fait, ses arguments et ce qu'elle renvoie. **Toute fonction doit être documentée!**
* Ensuite vient le corps de la fonction, qui calcule le plus grand des deux réels, et le renvoie avec le mot-clé `return`.

Pour utiliser cette fonction, on exécute un *appel* à celle-ci, et on récupère le résultat renvoyé dans une variable: `resultat=maximum(1,3.5)`. Ici, `maximum(...)` correspond à l'appel à la fonction, en lui donnant 2 valeurs pour ses arguments: *a* vaudra 1 et *b* vaudra 3,5. Cette fonction renvoie un résultat et il est stocké dans la variable `resultat`.
Pour savoir si une fonction est bien écrite, il faut la tester sur différentes valeurs. Pour cela, cette année, on écrira des tests (d'abord écrits dans les énoncés, ensuite vous les écrirez vous-mêmes) qui essayent de voir si la fonction renvoie la bonne valeurs sur quelques essais. Voici un exemple de test basique pour la fonction `maximum`:
```python
ok=True # pour l'instant, tout va bien
if maximum(1,2)!=2: ok=False # le max de 1 et 2 doit être 2
if maximum(-1,-2)!=-1: ok=False # avec les négatifs
if maximum(1.2,1.3)!=1.3: ok=False # dangereux
if ok:
    print("Le test de la fonction maximum a réussi")
else:
    print("Le test de la fonction maximum a échoué")
```

Évidemment, dans un vrai programme, il faudrait aussi tester les problèmes graves, du genre: comment réagit la fonction si *a* est un réel et *b* une chaîne de caractères?

*Remarque*: attention, si on définit des variables dans une fonction, elles n'existent que dans la fonction.

*Remarque*: pour éviter les confusions, **il est fortement déconseillé de déclarer dans une fonction une variable ayant le même nom que la fonction**.

## 9. Les listes

Python possède un type de données permettant de stocker une liste de valeurs; les listes sont définies par des crochets, les valeurs étant séparées par des virgules. Par exemple, `liste=[1,2,4,5,3]`. Chaque terme de cette liste est repéré par un *index*, il s'agit d'un entier qui repère la position de la case. Attention! **la première case a pour index 0**. Pour accéder à une case, on met son index entre crochet: ainsi `liste[2]` vaut 4, et l'instruction `liste[3]=6` modifiera la liste en `[1,2,4,6,3]`.

On peut ajouter un terme à la fin de la liste avec `append`. Ainsi si on écrit maintenant `liste.append(8)`, alors la liste vaudra désormais `[1,2,4,6,3,8]`. On peut ainsi construire une liste à partir de la liste vide `[]`.

On peut obtenir la longueur d'une liste avec `len(liste)`; attention, l'index du dernier terme est `len(liste)-1`.

Écrivez un programme qui demande à l'utilisateur un entier *n* et crée la liste contenant les *n* premiers termes de la suite définie par $u_0=0.5$ et $u_{n+1}=\frac{u_n+3}{2}$.

Les listes en Python sont très souples: on peut changer leur longueur en ajoutant des éléments, on peut aussi mélanger toutes sortes de variables: entiers, réels, chaînes de caractères,… C'est pratique, mais peu performant. Quand on a besoin de faire des calculs rapides, on utilise plutôt des *tableaux Numpy*, de longueur fixe, ne contenant qu'un seul type de données. On crée un tel tableau en important la bibliothèque `numpy` puis avec la commande `numpy.array([1,2,4,5,3])`; on peut aussi créer un tableau de zéros de longueur fixée: `numpy.zeros(100)`.

Le gros avantage des tableaux Numpy, c'est qu'on peut faire des calculs sur leurs termes (calcul vectoriel); par exemple si `t1` et `t2` sont deux tableaux de même dimension, alors `t1+t2` calcule la somme terme à terme des cases des deux tableaux, et `2*t1` double toutes les valeurs des cases du tableau.

Il existe une autre fonction très utile: `numpy.linspace(debut,fin,n)` qui crée un tableau de *n* valeurs également espacées entre le début et la fin. Testez par exemple le code suivant:
```python

print(x)
y=x**2
print(y)
```
