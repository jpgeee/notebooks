# Lentille simple

## Préliminaires

```python
def fprime(fpD,nD,n):
    """ calcule la distance focale d'une lentille à l'indice n
    connaissant sa distance focale à la raie D d'indice nD """
    return fpD*(nD-1)/(n-1)
```

```python
def nC(l):
    """ Calcule l'indice du verre Crown à la longueur d'onde l en nm """
    return 1.46+5200/l**2
def nF(l):
    """ Calcule l'indice du verre Flint à la longueur d'onde l en nm """
    return 1.62+18000/l**2
```

## Lentille simple

```python
import numpy as np
tab_l=np.linspace(400,800,100)
```

```python
tab_nC=nC(tab_l)
lD=589
nCD=nC(lD)
tab_fp=fprime(11,nCD,tab_nC)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,8))
plt.plot(tab_l,tab_fp)
plt.xlabel("lambda(nm)")
plt.ylabel("f'(cm)")
plt.title("Lentille simple")
plt.show()
```

```python
print(max(tab_fp)-min(tab_fp))
```

## Doublet achromatique

```python
import numpy as np
tab_l=np.linspace(400,800,100)
tab_nC=nC(tab_l)
tab_nF=nF(tab_l)
lD=589
tab_fp1=fprime(6,nC(lD),tab_nC)
tab_fp2=fprime(-12,nF(lD),tab_nF)
```

```python
def xImg(xObj,xLen,fp):
    """ calcule la position de l'objet, connaissant 
    la position de l'objet (xObj)
    la position de la lentille (xLen)
    la focale de la lentille (fp) """
    return xLen+((xObj-xLen)*fp)/(xObj-xLen+fp)
```

```python
def gamma(xObj,xLen,xImg):
    """ calcule le grandissement connaissant 
    la position de l'objet (xObj)
    la position de la lentille (xLen)
    la position de l'image (xImg) """
    return (xImg-xLen)/(xObj-xLen)
```

```python
xA=-1000000 # en cm
tab_xAp=xImg(xA,0,tab_fp1)
tab_xApp=xImg(tab_xAp,0.5,tab_fp2)
```

```python
plt.figure(figsize=(10,8))
plt.plot(tab_l,tab_xApp)
plt.xlabel("lambda(nm)")
plt.ylabel("f'(cm)")
plt.title("Doublet achromatique")
plt.show()
```

```python
print(max(tab_xApp)-min(tab_xApp))
```
